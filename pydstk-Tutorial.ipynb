Background
----------
Dynamical systems are an invaluable concept in many areas of science, such as 
control theory or physics. Over the past decade, they have also found numerous 
applications in computer vision. Especially, with the advent of **big** video data (e.g., youtube, vimeo), people have looked into ways of describing the dynamics as well as the appearance of those "visual processes" by generative models. This is appealing, since it allows to **synthesize** observations based on the estimated parameters of the dynamical system.

Estimating a dynamic texture model (i.e. System Identification)
----------------------------------

The (discrete-time) linear dynamical system (aka dynamic texture in the context of videos) can be written down as:

$x_{t+1} = \boldsymbol{A}\boldsymbol{x}_{t} + \boldsymbol{w}_t$

$y_{t} = \boldsymbol{C}\boldsymbol{x}_{t} + \boldsymbol{v}_t$

Both, state- and observation-evolvement are linear and the noise process (i.e., $w_t\sim \mathcal{N}(0,\boldsymbol{R})$ and $v_t \sim \mathcal{N}(0,\boldsymbol{Q}$) are zero-mean Gaussian (with different covariances). The task of **system identification** is to estimate the model parameters $(\boldsymbol{A},\boldsymbol{C},\boldsymbol{R},\boldsymbol{Q},x_0)$ from a set of observations $[\boldsymbol{y}_0-\bar{y},\ldots,\boldsymbol{y}_N-\bar{y}], \boldsymbol{y}_i \in \mathbb{R}^n$. In case of standard dynamic textures, the observations $\boldsymbol{y}_i$ are our video frames (and $\bar{y}$ is the mean, taken over all frames of the video).


We see that data is a (48 x 48) pixel video with 48 frames. **Note**: In the ASCII file this information needs to be included as the first line of the file (i.e., 48 48 48). Next, we will create a **Linear Dynamical System** (i.e., a dynamic texture in computer vision terminology) with 5 states:

````python
from dscore.system import LinearDS
````

````python
lds = LinearDS(5, False, False)
````

We can now run system identification to obtain the dynamic texture parameters. System identification is done via the suboptimal (SVD-based) approach that was originally proposed by Soatto et al.

````python
lds.check()
````

````python
lds.suboptimalSysID(data)
````

````python
lds.check()
````

As we have seen, ```check()``` returns ```False``` before we actually estimate the system parameters. This can be used to test if the dynamic texture model is ready for instance.

````python
lds._Ahat.shape
````

````python
lds._Chat.shape
````

So, the state-transition matrix **A** is a (5,5) matrix, the observation matrix **C** is a (48*48,48) matrix. We next check the stability of the system by looking at the **spectral radius**:

````python
import numpy as np
````

````python
eigval, eigvec = np.linalg.eig(lds._Ahat)
````

````python
spectral_radius = np.max(np.abs(eigval))
````

````python
print spectral_radius
````

As we can see, the spectral radius is less than 1, meaning that the system is stable. **Note**: This is NOT enforced in the estimation procedure at that point.

Clustering Dynamic Texture Models
---------------------------------
Clustering dynamic texture models is actually an important task when building bag-of-DT models, or when computational resources are limited and model comparision needs to be done w.r.t. a limited set of **representative** models from a database.

First, we are going to import a list of data file names. These files contain 2-D signals (sin,cos) with different frequencies, embedded in 24-D space.

Let's take a look at one 2-D signal:

````python
sig = np.genfromtxt("tests/data/clustering/sig0001.txt")
````

````python
plot(sig[0,:])
plot(sig[1,:])
````

Signals 0-49 were created using random frequencies in [0.9,1], signals 50-100 were created using random frequencies in [0.3, 0.4]. The signals are sampled at 80 linearly spaced points in [0,$6\pi$]. Some random noise ($\sigma=0.01$) was finally added to the embedded signal (in 24-dim. space).

````python
with open("tests/data/clustering.files") as fid: 
    content = fid.read().splitlines()
````

````python
content[0:5]
````

Then, we estimate a 8-state LDS for the content of each file:

````python
import os

models = []
for data_file in content:
    data = np.genfromtxt(os.path.join("tests/data/", data_file))
    
    lds = LinearDS(8, False, False)
    lds.suboptimalSysID(data)
    models.append(lds)
    
````

Let's create the pairwise distance matrix (between all the models) next:

````python
N = len(models)
D = np.zeros((N, N))

for i in range(N):
    for j in range(N):
        D[i,j] = ldsMartinDistance(models[i], models[j], 50)
````

Let's take a look at the distance matrix:

````python
imshow(D)
````

Looks not too bad. Let's run LDS clustering (with 2 cluster centers):

````python
eData, ids = LinearDS.cluster(D, 2)
````

````python
# first, we plot the data
plot(eData[0:49,0],eData[0:49,1], color='b', marker='+', linestyle='')
plot(eData[50:,0],eData[50:,1], color='r', marker='o', linestyle='')
# then, we plot the representatives
plot(eData[ids[0],0],eData[ids[0],1], marker='s', color='black', markersize=10)
plot(eData[ids[1],0],eData[ids[1],1], marker='s', color='green', markersize=10)
````

As we can see in this *toy example*, the LDS's are well separated and the cluster centers seem reasonable.
